```python
import numpy as np

names = ["김민수", "이서윤", "박지훈", "정예린", "최도현", "한지우", "유다인", "오승현", "강서영"]

# 3x3 배열로 변환
arr = np.array(names).reshape(3, 3)

# 3x5 배열을 "빈자리"로 초기화
result = np.full((3, 5), "빈자리")
# 이름을 [0, 2, 4] 위치에 배치
result[:, [0, 2, 4]] = arr

for r in result.tolist():
    print(r)
```

```python
import numpy as np

names = ["김민수", "이서윤", "박지훈", "정예린", "최도현", "한지우", "유다인", "오승현", "강서영"]

# 3x3 배열로 변환 후 "빈자리" 추가
arr = np.array(names).reshape(3, 3)  # 3x3 행렬로 변환
result = np.insert(arr, [1, 3], "빈자리", axis=1).tolist()  # 열 1, 3에 "빈자리" 삽입

print(result)
```
